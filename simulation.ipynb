```python
import random
import math
#cycle_count = list()
```

```python
def another_game_standalone():
        player_A = 4   
        player_B = 4
        pot = 2
        turn = 0
        game = True
        while game == True:
            turn += 1
            roll = random.randint(1,6)
            if roll == 2: 
                if (turn % 2) == 0:
                    player_B += pot
                    pot = 0
                    print("Player B got the pot!") 
                    print("A:",player_A,"B:",player_B,"Pot:",pot)
                else:
                    player_A += pot
                    pot = 0
                    print("Player A got the pot!")
                    print("A:",player_A,"B:",player_B,"Pot:",pot)

            elif roll == 3:
                if (turn % 2) == 0:
                    player_B += math.floor(pot/2) 
                    if pot > 0:
                        pot = math.ceil(pot/2)
                    else: 
                        pot = pot
                    print("Player B got half the pot")
                    print("A:",player_A,"B:",player_B,"Pot:",pot)
                else:
                    player_A += math.floor(pot/2)
                    if pot > 0:
                        pot = math.ceil(pot/2)
                    else: 
                        pot = pot
                    print("Player A got half the pot")
                    print("A:",player_A,"B:",player_B,"Pot:",pot)

            elif roll == 4 or roll == 5 or roll == 6:
                if (turn % 2) == 0 and player_B >= 1:
                    pot += 1
                    player_B -= 1
                    print("Player B put one in the pot")
                    print("A:",player_A,"B:",player_B,"Pot:",pot)
                elif (turn %2) == 1 and player_A >= 1:
                    pot += 1
                    player_A -= 1
                    print("Player A put one in the pot")
                    print("A:",player_A,"B:",player_B,"Pot:",pot)
                elif (turn %2) == 0 and player_B == 0: 
                    game = False
                    print("Game over, player B has no coins left to give!")
                    print("A:",player_A,"B:",player_B,"Pot:",pot)
                    print("The game lasted", math.ceil(turn/2), "cycles.")
                elif (turn % 2) == 1 and player_A == 0:
                    game = False
                    print("Game over, player A has no coins left to give!")
                    print("A:",player_A,"B:",player_B,"Pot:",pot)
                    print("The game lasted", math.ceil(turn/2), "cycles.")
```

```python
def another_game_1():
        player_A = 4   
        player_B = 4
        pot = 2
        turn = 0
        game = True
        while game == True:
            turn += 1 #Odd turns are Player A, even turns are Player B
            cycle = math.ceil(turn/2)
            roll = random.randint(1,6)
            if roll == 2: 
                if (turn % 2) == 0: #Player B gets the pot
                    player_B += pot
                    pot = 0
                else: #Player A gets the pot
                    player_A += pot
                    pot = 0                 
            elif roll == 3:
                if (turn % 2) == 0: #Player B gets the lesser half of the pot, as long as the pot is not empty
                    player_B += math.floor(pot/2) 
                    if pot > 0:
                        pot = math.ceil(pot/2)
                    else: 
                        pot = 0
                else: #Player A gets the lesser half of the pot, as long as the pot is not empty
                    player_A += math.floor(pot/2)
                    if pot > 0:
                        pot = math.ceil(pot/2)
                    else: 
                        pot = 0
            elif roll == 4 or roll == 5 or roll == 6: 
                if (turn % 2) == 0 and player_B >= 1: #If player B has coins, they put one in the pot
                    pot += 1
                    player_B -= 1
                elif (turn %2) == 1 and player_A >= 1: #If player A has coins, they put one in the pot
                    pot += 1
                    player_A -= 1
                elif (turn %2) == 0 and player_B == 0: #If player B has no coins, then they lose and the game ends
                    game = False
                    return cycle
                elif (turn % 2) == 1 and player_A == 0: #If player A has no coins, then they lose and the game ends
                    game = False
                    return cycle
```

```python
another_game_1()
```

```python
###Replicate game to see the average number of cycles it lasts 
r = 100000
#total_cycles = 0
cycles_list = list()
for i in range(r):
    #total_cycles += another_game_1()
    cycles_list.append(another_game_1())
avg_cycles = sum(cycles_list) / r
print("Average number of cycles:", avg_cycles)
```

```python
###See number of occurrences of each cycle length and percentages.

for i in range(min(cycles_list),max(cycles_list)):
    print(i, "occurs", cycles_list.count(i), " times which is", round(cycles_list.count(i)/len(cycles_list)*100,2), "percent.")
```

```python
###Plot Histogram to see frequency distribution
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
cycle_count = np.array(cycles_list)
cycle_count1 = pd.Series(cycles_list)
    
#cycle_count1.plot.hist(grid = True, bins= 30, rwidth = 0.9, color = '#607c8e') #This works but title 
plt.hist(cycle_count1, bins = 30)
plt.title('Histogram of Cycle Counts')
plt.xlabel('Number of Cycles a Game Lasts')
plt.ylabel('Frequency')
plt.grid(axis = 'y', alpha = 0.75)
```

```python
###GAME STATS FUNCTION WITH GAME INCLUDED

def game_stats(x):
    cycle_count = list()
    a_loss = list()
    b_loss = list()
    two_loss = list()
    three_loss = list()
    fourfivesix_loss = list()
    def another_game():
        player_A = 4   
        player_B = 4
        pot = 2
        turn = 0
        cycle = 0 
        game = True
        while game == True:
            turn += 1
            roll = random.randint(1,6)
            if roll == 2: 
                if (turn % 2) == 0:
                    player_B += pot
                    pot = 0
                    #print("Player B got the pot!") 
                else:
                    player_A += pot
                    pot = 0
                    #print("Player A got the pot!")

            elif roll == 3:
                if (turn % 2) == 0:
                    player_B += math.floor(pot/2) 
                    if pot > 0:
                        pot = math.ceil(pot/2)
                    else: 
                        pot = 0
                    #print("Player B got half the pot")
                else:
                    player_A += math.floor(pot/2)
                    if pot > 0:
                        pot = math.ceil(pot/2)
                    else: 
                        pot = 0
                    #print("Player A got half the pot")

            elif roll == 4 or roll == 5 or roll == 6:
                if (turn % 2) == 0 and player_B >= 1:
                    pot += 1
                    player_B -= 1
                    #print("Player B put one in the pot")
                elif (turn %2) == 1 and player_A >= 1:
                    pot += 1
                    player_A -= 1
                    #print("Player A put one in the pot")
                elif (turn %2) == 0 and player_B == 0: 
                    game = False
                    fourfivesix_loss.append(1)
                    b_loss.append(1)
                elif (turn % 2) == 1 and player_A == 0:
                    game = False
                    fourfivesix_loss.append(1)
                    a_loss.append(1)
                    
        #print("Game", i, "over after", turn, "turns.") #count statistics of what ends the game
        cycle = math.ceil(turn/2) 
        cycle_count.append(cycle)
    i = 1
    while i <= x:
        another_game()
        i += 1
    print("After", len(cycle_count), "games, the average game length was", sum(cycle_count)/len(cycle_count), "cycles.")
    print(cycle_count)
    print(round(len(a_loss) / len(cycle_count)*100,1), "% of games were lost by player A.")
    print(round(len(b_loss) / len(cycle_count)*100,1), "% of games were lost by player B.")
    print(round(len(two_loss) / len(cycle_count)*100,1), "% of games were lost on a 2 roll.")
    print(round(len(three_loss) / len(cycle_count)*100,1), "% of games were lost on a 3 roll.")
    print(round(len(fourfivesix_loss) / len(cycle_count)*100,1), "% of games were lost on a 4, 5, or 6 roll.")
    
```
